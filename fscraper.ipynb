```python
import pandas as pd
import csv
import requests as r
from bs4 import BeautifulSoup as bs


def data_rows(product=str):
    try:
        url=f'https://www.flipkart.com/search?q={product}'
        page=r.get(url)


        soup=bs(page.content,'html.parser')
        products=soup.find_all(class_="_13oc-S")


        for product in products:
            link=product.find(class_="_1fQZEK")
            url=link['href']
            name=product.find(class_="_4rR01T")
            rating=product.find(class_="_3LWZlK")
            price=product.find(class_="_30jeq3 _1_WHN1")

            n=name.text
            if not rating:
                m=None
            else:m=(rating.text)
            p=price.text
            l='https://www.flipkart.com'+url

            yield [n,m,p,l]
    except:
        return 'try different product category'

    
file = open('product_info.csv',mode='w+')
writer=csv.writer(file,delimiter=',')
writer.writerow(['name','rating','price','product link'])
result=list(data_rows('moto'))
writer.writerows(result)
file.close()
```

```python
def data_columns(product=str):
    
    url=f'https://www.flipkart.com/search?q={product}'
    page=r.get(url)
    
    try:
        soup=bs(page.content,'html.parser')
        products=soup.find_all(class_="_13oc-S")

        name_col=['name']
        rating_col=['rating']
        price_col=['price']
        link_col=['link']

        for product in products:

            link=product.find(class_="_1fQZEK")
            url=link['href']
            l='https://www.flipkart.com'+url

            name=product.find(class_="_4rR01T")

            rating=product.find(class_="_3LWZlK")
            if not rating:
                m='None'
            else:m=rating.text

            price=product.find(class_="_30jeq3 _1_WHN1")


            name_col=name_col+[name.text]
            rating_col=rating_col+[m]
            price_col=price_col+[price.text]
            link_col=link_col+[l]
            
        return [name_col,rating_col, price_col,link_col]   
    
    except:
        return 'try different category of product like phones or TVs'
    
            
```

```python
def create_csv(f):
    def wrap():
        file = open('yproduct_info.csv',mode='w+')
        writer=csv.writer(file,delimiter=',')
        writer.writerow(['name','rating','price','product link'])
        result=list(f())
        for i in result:
            writer.writerow(i)
    return wrap()
```
